**Exercício 01: Automação de Arquivos**



image.png

```python
import os
import shutil
```

### Defina os diretórios de origem e destino

- `source_dir`: O diretório onde seus arquivos bagunçados estão localizados.
- `destination_dir`: O diretório onde as novas pastas organizadas serão criadas.

```python
# Defina o diretório de origem (onde estão os arquivos que você quer organizar)
source_dir = "./arquivos_baguncados"

# Defina o diretório de destino (onde os arquivos organizados serão movidos)
destination_dir = "./arquivos_organizados"

# Cria os diretórios se não existirem
os.makedirs(source_dir, exist_ok=True)
os.makedirs(destination_dir, exist_ok=True)

print(f"Diretório de origem: {os.path.abspath(source_dir)}")
print(f"Diretório de destino: {os.path.abspath(destination_dir)}")
```

### Crie 15 arquivos de exemplo para testar o programa

Estes arquivos serão usados para demonstrar como o script organiza diferentes tipos de arquivos.

```python
# Criar 15 arquivos de exemplo no diretório de origem
open(os.path.join(source_dir, "documento1.pdf"), "w").close()
open(os.path.join(source_dir, "documento2.pdf"), "w").close()
open(os.path.join(source_dir, "imagem1.jpg"), "w").close()
open(os.path.join(source_dir, "imagem2.jpg"), "w").close()
open(os.path.join(source_dir, "foto1.png"), "w").close()
open(os.path.join(source_dir, "foto2.png"), "w").close()
open(os.path.join(source_dir, "nota1.txt"), "w").close()
open(os.path.join(source_dir, "nota2.txt"), "w").close()
open(os.path.join(source_dir, "relatorio1.docx"), "w").close()
open(os.path.join(source_dir, "relatorio2.docx"), "w").close()
open(os.path.join(source_dir, "planilha.xlsx"), "w").close()
open(os.path.join(source_dir, "apresentacao.pptx"), "w").close()
open(os.path.join(source_dir, "musica.mp3"), "w").close()
open(os.path.join(source_dir, "video.mp4"), "w").close()
open(os.path.join(source_dir, "script.py"), "w").close()

print(f"15 arquivos de exemplo criados em '{source_dir}'")
```

### Lógica principal para organizar os arquivos

1. Itera sobre cada item no `source_dir`.
2. Verifica se é um arquivo (e não um diretório).
3. Obtém a extensão do arquivo.
4. Cria uma subpasta no `destination_dir` com o nome da extensão (ex: `pdf`, `jpg`).
5. Move o arquivo para a subpasta correspondente.

```python
for filename in os.listdir(source_dir):
    # Constrói o caminho completo do arquivo
    file_path = os.path.join(source_dir, filename)

    # Verifica se é realmente um arquivo e não um diretório
    if os.path.isfile(file_path):
        # Obtém a extensão do arquivo (ex: '.pdf', '.jpg')
        file_extension = os.path.splitext(filename)[1].lower() # .lower() para padronizar

        # Se o arquivo não tiver extensão, pode-se criar uma pasta 'sem_extensao' ou ignorar
        if not file_extension:
            print(f"Ignorando arquivo sem extensão: {filename}")
            continue

        # Remove o ponto da extensão para usar como nome da pasta (ex: 'pdf', 'jpg')
        folder_name = file_extension[1:]

        # Constrói o caminho para a nova pasta de destino
        destination_folder = os.path.join(destination_dir, folder_name)

        # Cria a pasta de destino se ela não existir
        os.makedirs(destination_folder, exist_ok=True)

        # Constrói o caminho completo para onde o arquivo será movido
        new_file_path = os.path.join(destination_folder, filename)

        # Move o arquivo do diretório de origem para o diretório de destino
        shutil.move(file_path, new_file_path)
        print(f"Movido '{filename}' para '{folder_name}/'")

print("Organização de arquivos concluída!")
```

### Verificar a organização

Você pode executar esta célula para ver como os arquivos foram organizados no `destination_dir`.

```python
print(f"Conteúdo do diretório organizado '{destination_dir}':")
for root, dirs, files in os.walk(destination_dir):
    level = root.replace(destination_dir, '').count(os.sep)
    indent = ' ' * 4 * (level)
    print(f'{indent}{os.path.basename(root)}/')
    subindent = ' ' * 4 * (level + 1)
    for f in files:
        print(f'{subindent}{f}')
```

**Exercício 02: Consulta simples a APIs**

```python
import requests

def consultar_cep(cep):
    """Consulta o ViaCEP para obter informações de endereço."""
    # Remove qualquer caractere não numérico do CEP
    cep = ''.join(filter(str.isdigit, cep))

    if len(cep) != 8:
        return {"erro": "CEP deve conter 8 dígitos."}

    url = f"https://viacep.com.br/ws/{cep}/json/"
    try:
        response = requests.get(url)
        response.raise_for_status()  # Levanta um erro para status de HTTP ruins (4xx ou 5xx)
        data = response.json()

        if "erro" in data:
            return {"erro": "CEP não encontrado."}
        else:
            return data
    except requests.exceptions.RequestException as e:
        return {"erro": f"Erro na conexão: {e}"}
    except ValueError:
        return {"erro": "Resposta inválida da API."}

def formatar_endereco(data):
    """Formata as informações do endereço de forma legível."""
    if "erro" in data:
        return data["erro"]
    else:
        return (
            f"CEP:        {data.get('cep', 'N/A')}\n"
            f"Logradouro: {data.get('logradouro', 'N/A')}\n"
            f"Complemento: {data.get('complemento', 'N/A')}\n"
            f"Bairro:     {data.get('bairro', 'N/A')}\n"
            f"Localidade: {data.get('localidade', 'N/A')} - {data.get('uf', 'N/A')}\n"
            f"IBGE:       {data.get('ibge', 'N/A')}\n"
            f"DDD:        {data.get('ddd', 'N/A')}"
        )

# Exemplo de uso:
if __name__ == "__main__":
    while True:
        cep_input = input("Digite o CEP para consulta (ex: 01001000) ou 'sair' para finalizar: ")
        if cep_input.lower() == 'sair':
            print("Saindo da consulta de CEP.")
            break

        info_endereco = consultar_cep(cep_input)
        print("\n--- Informações do Endereço ---")
        formatted_output = formatar_endereco(info_endereco)

        if "erro" in info_endereco:
            print(f"OPÇÃO INVÁLIDA: {formatted_output}")
            # Optionally, ask if they want to try again, but the loop handles it.
            # print("Por favor, tente novamente.")
        else:
            print(formatted_output)
            break # Exit the loop if a valid CEP is found
```

**Exercício 03: Sistema de Notificações**

image.png

```python
import time
import random

# --- Configurações --- Foster free thinking, if the solution is complicated, simplify it. Think step by step.
CRITICAL_TEMPERATURE_THRESHOLD = 30  # Graus Celsius
CHECK_INTERVAL_SECONDS = 5          # Frequência de verificação em segundos

# --- Funções de Alerta ---
def send_console_alert(message):
    """Envia um alerta para o console."""
    print(f"[ALERTA CRÍTICO] {message}")

def simulate_email_alert(recipient, subject, body):
    """Simula o envio de um e-mail de alerta."""
    print(f"\n--- E-mail Simulado Enviado para: {recipient} ---")
    print(f"Assunto: {subject}")
    print(f"Corpo:\n{body}")
    print("-------------------------------------------\n")

# --- Programa Autônomo ---
print("Iniciando monitoramento autônomo do sistema...")
print(f"Temperatura crítica definida em: > {CRITICAL_TEMPERATURE_THRESHOLD}°C")
print(f"Verificando a cada {CHECK_INTERVAL_SECONDS} segundos.\n")

alert_triggered = False # Flag para garantir que o alerta crítico só seja disparado uma vez até a condição normalizar

while True:
    # Simula a leitura de uma temperatura de um sensor
    # Para demonstração, a temperatura vai subir gradualmente ou aleatoriamente para atingir a condição crítica
    # Em um sistema real, aqui você leria de um sensor físico ou API
    current_temperature = random.uniform(20.0, 35.0) # Gera uma temperatura entre 20 e 35 para simular variações

    print(f"Verificando... Temperatura atual: {current_temperature:.2f}°C")

    if current_temperature > CRITICAL_TEMPERATURE_THRESHOLD:
        if not alert_triggered: # Dispara o alerta apenas se ainda não foi disparado
            alert_message = f"Superaquecimento detectado! Temperatura: {current_temperature:.2f}°C (Limiar: {CRITICAL_TEMPERATURE_THRESHOLD}°C)"

            # Alerta no console
            send_console_alert(alert_message)

            # Alerta via e-mail simulado
            simulate_email_alert(
                recipient="admin@empresa.com",
                subject="ALERTA CRÍTICO: Superaquecimento do Sistema!",
                body=f"Prezado Administrador,\n\nFoi detectado um superaquecimento no sistema.\nDetalhes: {alert_message}\nPor favor, investigue imediatamente."
            )
            alert_triggered = True # Marca que o alerta foi disparado
        else:
            print("Alerta crítico já disparado, aguardando normalização...")
    else:
        if alert_triggered: # Se a temperatura voltou ao normal após um alerta
            print("A condição crítica foi resolvida. Temperatura normalizada.")
            alert_triggered = False # Reseta a flag para permitir novos alertas
        else:
            print("Temperatura normal.")

    time.sleep(CHECK_INTERVAL_SECONDS)
```
